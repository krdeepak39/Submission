```python
from bs4 import BeautifulSoup as bs
import requests
import csv
```

```python
r =  requests.get("https://greenstocknews.com/stocks/electric-vehicle-stocks")
soup = bs(r.content)
contents = soup.prettify()
```

```python
print(contents)
```

```python
soup.find_all(class_="table-header")
```

```python
soup.find_all(class_ = "table-content stockslist-items")
```

```python
name = soup.find_all(class_="stock-title", )
```

```python
Value =soup.find_all(class_="last-trade" )
change =soup.find_all(class_="change-value")
vol = soup.find_all(class_="volume-average")
share = soup.find_all(class_="totalShares")
MP=soup.find_all(class_="market-cap")

```

```python
Company = []
last_trade = []
Changes = []
Volume = []
Share_out = []
Market_Cap = []

for i in name:
    Company.append(i.text)
for j in Value:
    last_trade.append(j.text)
for k in change:
    Changes.append(k.text)
for l in vol:
    Volume.append(l.text)
for m in share:
    Share_out.append(m.text)
for n in MP:
    Market_Cap.append(n.text)

    
```

```python
print((Company, last_trade, Changes, Volume, Share_out, Market_Cap))
```

```python
with open('submission.csv', 'w', newline='') as f:
    
    f.write("Company, Last Trade, Changes, Volume, Share Outstanding, Market Cap\n")

   
    for row in zip(Company, last_trade[1], Changes, Volume, Share_out, Market_Cap[1]):
        f.write(','.join(map(str, row)) + '\n')
```
